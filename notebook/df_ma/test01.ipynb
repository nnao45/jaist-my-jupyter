```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_historical = pd.read_csv('./quote_utf8.csv')
df_historical
```

```python
ma24 = df_historical.USD.rolling(24).mean()
diff_ma24 = (df_historical.USD - ma24) / ma24 * 100
diff_ma24
```

```python
fig, ax = plt.subplots()
ax.grid()

ax.plot(diff_ma24)
fig.tight_layout()
```
